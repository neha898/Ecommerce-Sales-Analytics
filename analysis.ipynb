```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../dataset/samplesuperstore.csv")

df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df.describe()
```

```python
df.drop_duplicates(inplace=True)
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(kind='bar')

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

```python
category_profit = df.groupby('Category')['Profit'].sum()

category_profit.plot(kind='bar')

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Region-wise Sales")

plt.ylabel("")

plt.show()
```

```python
top_states = df.groupby('State')['Sales'].sum() \
               .sort_values(ascending=False) \
               .head(10)

top_states.plot(kind='bar')

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")

plt.show()
```

```python
sns.scatterplot(
    x='Discount',
    y='Profit',
    data=df
)

plt.title("Discount vs Profit")

plt.show()
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum()

segment_sales.plot(kind='bar')

plt.title("Sales by Segment")
plt.xlabel("Segment")
plt.ylabel("Sales")

plt.show()
```

```python
ship_mode = df['Ship Mode'].value_counts()

ship_mode.plot(kind='bar')

plt.title("Shipping Mode Usage")
plt.xlabel("Ship Mode")
plt.ylabel("Orders")

plt.show()
```

```python
top_subcat = df.groupby('Sub-Category')['Profit'].sum() \
               .sort_values(ascending=False) \
               .head(10)

top_subcat.plot(kind='bar')

plt.title("Top Sub-Categories by Profit")
plt.xlabel("Sub-Category")
plt.ylabel("Profit")

plt.show()
```

```python
loss_subcat = df.groupby('Sub-Category')['Profit'].sum() \
                .sort_values()

loss_subcat.head(10).plot(kind='bar')

plt.title("Loss-Making Sub-Categories")
plt.xlabel("Sub-Category")
plt.ylabel("Profit")

plt.show()
```
